```python
#make 60 directories
for i in range(60):
    if os.path.isdir(os.path.join(basedir+'/datasets',str(i)))==False:
        os.mkdir(path=os.path.join(basedir+'/datasets',str(i)))
```

```python
import shutil
def move_files(source_path,des_path,label):
    print('src=',source_path)
    print('des=',des_path)
    for file in os.listdir(source_path):
        
        filename = os.fsdecode(file)
        path = os.path.join(source_path,filename)
        if os.path.exists(os.path.join(des_path,filename)):
            print('file exists')
            continue
        dest = shutil.move(path,des_path)
    print('final dest was ',dest)
```

```python
move_files(os.path.join(basedir+'/EkushImg','male/10'),os.path.join(basedir+'/datasets','0'),0)
move_files(os.path.join(basedir+'/EkushImg','female/10'),os.path.join(basedir+'/datasets','0'),0)
```

```python
#get all data from EkushImg

def organize_datasets(source_dir,dest_dir):
    
    lstdir = os.listdir(source_dir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=int(lstdir[i])
    lstdir=sorted(lstdir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=str(lstdir[i])

    print(lstdir)
    i=0
    for dir_ in lstdir:
        path = os.path.join(source_dir,dir_)
        print('dir=',path)
        if os.path.isdir(path):
            dest =  os.path.join(dest_dir,str(i))
            print('dest=',dest)
            move_files(path,dest,label=i)
            i+=1
    
```

```python
#populate datasets with Ekush data
organize_datasets(basedir+'/EkushImg/male',basedir+'/datasets')
organize_datasets(basedir+'/EkushImg/female',basedir+'/datasets')
```
