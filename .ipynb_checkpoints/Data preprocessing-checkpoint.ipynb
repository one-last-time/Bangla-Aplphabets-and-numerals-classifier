```python
import shutil
def move_files(source_path,des_path,label):
    print('src=',source_path)
    print('des=',des_path)
    for file in os.listdir(source_path):
        
        filename = os.fsdecode(file)
        path = os.path.join(source_path,filename)
        if os.path.exists(os.path.join(des_path,filename)):
            print('file exists')
            continue
        dest = shutil.move(path,des_path)
    print('final dest was ',dest)
```

```python
#get all data from EkushImg

def organize_datasets(source_dir,dest_dir):
    
    lstdir = os.listdir(source_dir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=int(lstdir[i])
    lstdir=sorted(lstdir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=str(lstdir[i])

    print(lstdir)
    i=0
    for dir_ in lstdir:
        path = os.path.join(source_dir,dir_)
        print('dir=',path)
        if os.path.isdir(path):
            dest =  os.path.join(dest_dir,str(i))
            print('dest=',dest)
            move_files(path,dest,label=i)
            i+=1
    
```

```python
#populate datasets with Ekush data
organize_datasets(basedir+'/EkushImg/male',basedir+'/training')
organize_datasets(basedir+'/EkushImg/female',basedir+'/training')
```

```python
#get all data from BanglaLekha-Isolated
organize_datasets(basedir+'/BanglaLekha-Isolated/Images',basedir+'/training')
```

```python
def move_validation_files(source_path,des_path,label):
    print('src=',source_path)
    print('des=',des_path)
    x= int(len(os.listdir(source_path))*0.2)
    print('x=',x)
    for file in os.listdir(source_path):
        if x<=0:
            break
        filename = os.fsdecode(file)
        path = os.path.join(source_path,filename)
        if os.path.exists(os.path.join(des_path,filename)):
            print('file exists')
            continue
        dest = shutil.move(path,des_path)
        x-=1
    print('final dest was ',dest)
```

```python
def organize_validation_datasets(source_dir,dest_dir):
    
    lstdir = os.listdir(source_dir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=int(lstdir[i])
    lstdir=sorted(lstdir)
    for i,item in enumerate(lstdir): 
        lstdir[i]=str(lstdir[i])

    print(lstdir)
    i=0
    for dir_ in lstdir:
        path = os.path.join(source_dir,dir_)
        print('dir=',path)
        if os.path.isdir(path):
            dest =  os.path.join(dest_dir,str(i))
            print('dest=',dest)
            move_validation_files(path,dest,label=i)
            i+=1
```

```python
organize_validation_datasets(os.path.join(basedir,'training'),os.path.join(basedir,'validation'))
```
